```python
import pandas as pd
import numpy as np
```

```python
Bangalore=pd.read_csv(r"C:\Users\prave\OneDrive\Documents\GUVI\car dheko\bangalore_cars.csv")
Chennai=pd.read_csv(r"C:\Users\prave\OneDrive\Documents\GUVI\car dheko\chennai_cars.csv")                   
Delhi=pd.read_csv(r"C:\Users\prave\OneDrive\Documents\GUVI\car dheko\delhi_cars.csv")
Hyderabad=pd.read_csv(r"C:\Users\prave\OneDrive\Documents\GUVI\car dheko\hyderabad_cars.csv")
Jaipur=pd.read_csv(r"C:\Users\prave\OneDrive\Documents\GUVI\car dheko\jaipur_cars.csv")
Kolkata=pd.read_csv(r"C:\Users\prave\OneDrive\Documents\GUVI\car dheko\jaipur_cars.csv")
```

```python
Final_report=pd.concat([Bangalore,Chennai,Delhi,Hyderabad,Jaipur,Kolkata],axis=0)
Final_report
```

```python
Final_report.isnull()
Final_report.shape
```

```python
necessary_columns = [
    'ft',                 
    'bt',                 
    'transmission',       
    'ownerNo',            
    'oem',                
    'model',              
    'modelYear',          
    'price', 
    'Insurance Validity', 
    'Kms Driven',         
    'Mileage',            
    'Engine',             
    'Seats',              
    'Color',              
    'City'                
]
```

```python
Final_report=Final_report[necessary_columns]
Final_report
```

```python
Final_report=Final_report.rename({"ft": "Fuel_Type", "bt": "Body_Type", "oem": "Brand", "Insurance Validity": "Insurance_Validity","Kms Driven":"Kms_Driven","Gear Box": "Gear_Box","Drive Type": "Drive_Type"}, axis='columns')
Final_report
```

```python
def price(amount):
    amount = str(amount).replace("₹", "").replace(",", "").strip().lower()
    if 'crore' in amount:
        return round(float(amount.replace("crore", "").strip()) * 1e7, 2)
    elif 'lakh' in amount:
        return round(float(amount.replace("lakh", "").strip()) * 1e5, 2)
    else:
        try:
            return round(float(amount), 2)  # Changed 'price' to 'amount'
        except ValueError:
            return None

```

```python
# Apply the function to the 'price' column
Final_report['price'] = Final_report['price'].apply(price)

# Convert the 'price' column to float (this step is now redundant as 'price' is already float)
Final_report['price'] = Final_report['price'].astype(float)
Final_report['price']
```

```python
Final_report['Body_Type'] = Final_report['Body_Type'].fillna(Final_report['Body_Type'].mode()[0])
Final_report['Body_Type']
```

```python
Final_report['Body_Type'].isnull()
```

```python
Final_report['Insurance_Validity']=Final_report['Insurance_Validity'].fillna(Final_report['Insurance_Validity'].mode()[0])
```

```python
Final_report['Kms_Driven']=Final_report['Kms_Driven'].str.replace("Kms","").str.replace(",","").str.strip()
Final_report['Kms_Driven']=Final_report['Kms_Driven'].astype(float)
```

```python
Final_report['Kms_Driven']=Final_report['Kms_Driven'].fillna(Final_report['Kms_Driven'].mean())
```

```python
Final_report['Mileage']=Final_report['Mileage'].str.extract(r'([\d\.]+)').astype(float)
Final_report['Mileage']=Final_report['Mileage'].fillna(Final_report['Mileage'].mean())
```

```python
Final_report['Engine']=Final_report['Engine'].str.extract(r'([\d\.]+)').astype(float)
Final_report['Engine']=Final_report['Engine'].fillna(Final_report['Engine'].mean()).astype(int)
```

```python
Final_report['Seats']=Final_report['Seats'].str.extract(r'([\d\.]+)').astype(float)
Final_report['Seats']=Final_report['Seats'].fillna(Final_report['Seats'].mean()).astype(int)
```

```python
Final_report['Color'] = Final_report['Color'].fillna(Final_report['Color'].mode()[0])
```

```python
Final_report.drop_duplicates(inplace=True)
```
